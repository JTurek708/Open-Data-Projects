```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
# Import updated files - 2020 thru 2023
filenames = [f'{year}{month:02d}-divvy-tripdata.csv' for year in range(2020,2024) for month in range(1,13)]
```

```python
filenames
```

```python
# Adjust for files that don't follow the pattern
filenames.remove('202001-divvy-tripdata.csv')
filenames.remove('202002-divvy-tripdata.csv')
filenames.remove('202003-divvy-tripdata.csv')
filenames.remove('202209-divvy-tripdata.csv')
```

```python
# Append new files
filenames.append('Divvy_Trips_2020_Q1.csv')
filenames.append('202209-divvy-publictripdata.csv')
```

```python
# Remove filenames beyond October 2023
filenames = [f for f in filenames if not (f.startswith('202311') or f.startswith('202312'))]
```

```python
# Read csv file and store in list
df_list = []
for filename in filenames:
    try:
        df = pd.read_csv(filename)
        df_list.append(df)
    except FileNotFoundError:
        print(f"File Not Found: {filename}")

# Concat all Dataframes into one
updated_divvy_data = pd.concat(df_list, ignore_index=True)
```

```python
updated_divvy_data.head()
updated_divvy_data.tail()
updated_divvy_data.shape
```

```python
updated_divvy_data.head()
```

```python
updated_divvy_data['started_at'] = pd.to_datetime(updated_divvy_data['started_at'])
sorted_df = updated_divvy_data.sort_values(by='started_at', ascending=True)
```

```python
sorted_df.head()
```

```python
sorted_df.reset_index(drop=True, inplace=True)
sorted_df.head()
```

```python
# Create list of new column names
new_cols = ['TRIP ID', 'RIDE TYPE', 'START TIME', 'STOP TIME', 'FROM STATION NAME',
            'FROM STATION ID', 'TO STATION NAME', 'TO STATION ID', 'FROM LATITUDE',
            'FROM LONGITUDE', 'TO LATITUDE', 'TO LONGITUDE', 'USER TYPE']

sorted_df.columns = new_cols
sorted_df.head()
```

```python
sorted_df.drop('RIDE TYPE', axis=1, inplace=True)
```

```python
import numpy as np
sorted_df['BIKE ID'] = np.nan
sorted_df['TRIP DURATION'] = np.nan
sorted_df['GENDER'] = np.nan
sorted_df['BIRTH YEAR'] = np.nan
sorted_df['TO LOCATION'] = np.nan
sorted_df.head()
```

```python
# Replace USER TYPE values
replace_values = {'casual': 'Customer', 'member': 'Subscriber'}
sorted_df['USER TYPE'] = sorted_df['USER TYPE'].replace(replace_values)
```

```python
sorted_df.head()
```

```python
# Convert start time and stop time to pd datetime
sorted_df['START TIME'] = pd.to_datetime(sorted_df['START TIME']).dt.strftime('%m/%d/%Y %I:%M:%S %p')
sorted_df['STOP TIME'] = pd.to_datetime(sorted_df['STOP TIME']).dt.strftime('%m/%d/%Y %I:%M:%S %p')
```

```python
sorted_df.head()
```

```python
sorted_df['START TIME'] = pd.to_datetime(sorted_df['START TIME'], format="%m/%d/%Y %I:%M:%S %p")
sorted_df['STOP TIME'] = pd.to_datetime(sorted_df['STOP TIME'], format="%m/%d/%Y %I:%M:%S %p")
```

```python
sorted_df.head()
```

```python
# Break out Month Day Year and Time for starts and stops
sorted_df['Start Month'] = sorted_df['START TIME'].dt.month
sorted_df['Start Day'] = sorted_df['START TIME'].dt.day
sorted_df['Start Year'] = sorted_df['START TIME'].dt.year
sorted_df['Start Time'] = sorted_df['START TIME'].dt.strftime('%H%M')

sorted_df['Stop Month'] = sorted_df['STOP TIME'].dt.month
sorted_df['Stop Day'] = sorted_df['STOP TIME'].dt.day
sorted_df['Stop Year'] = sorted_df['STOP TIME'].dt.year
sorted_df['Stop Time'] = sorted_df['STOP TIME'].dt.strftime('%H%M')

sorted_df.head()
```

```python
sorted_df.shape
```

```python
sorted_df.columns
```

```python
# Add weekday or weekend value
import datetime
def is_weekday(year, month, day):
    # Create datetime object
    date=datetime.date(year, month, day)
    # Check if it's a weekday
    return "Weekday" if date.weekday() < 5 else "Weekend"

sorted_df['Day Type'] = sorted_df.apply(lambda row: is_weekday(row['Start Year'], row['Start Month'], row['Start Day']),axis=1)
```

```python
sorted_df.head()
```

```python
sorted_df['FROM LOCATION'] = np.nan
```

```python
# re-order col names
new_col_order = ['TRIP ID', 'START TIME', 'STOP TIME', 'BIKE ID', 'TRIP DURATION',
                 'FROM STATION ID', 'FROM STATION NAME', 'TO STATION ID', 'TO STATION NAME',
                 'USER TYPE', 'GENDER', 'BIRTH YEAR', 'FROM LATITUDE', 'FROM LONGITUDE',
                 'FROM LOCATION', 'TO LATITUDE', 'TO LONGITUDE', 'TO LOCATION', 'Start Month',
                 'Start Day', 'Start Year', 'Start Time', 'Stop Month', 'Stop Day', 'Stop Year',
                 'Stop Time', 'Day Type']

sorted_df = sorted_df[new_col_order]
```

```python
sorted_df.head()
```

```python
sorted_df.shape
```

```python
# Calculate TRIP DURATION in seconds
sorted_df['TRIP DURATION'] = (sorted_df['STOP TIME'] - sorted_df['START TIME']).dt.total_seconds()
sorted_df.head()
```

```python
sorted_df['TRIP DURATION'] = sorted_df['TRIP DURATION'].astype(int)
sorted_df.head()
```

```python
sorted_df['Rider Age'] = np.nan
```

```python
# Write new data to csv
sorted_df.to_csv('updated_divvy_data.csv', index=False)
```
